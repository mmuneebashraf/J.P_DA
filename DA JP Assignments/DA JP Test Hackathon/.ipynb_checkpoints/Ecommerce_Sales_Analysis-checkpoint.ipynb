# Ecommerce Sales Analysis
This notebook contains the complete improved analysis, preprocessing, and visualization for the Ecommerce dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('MhZsQF.csv')

# Convert date column
df['date'] = pd.to_datetime(df['date'])
df['Month'] = df['date'].dt.month_name()

# Drop irrelevant columns
df.drop(['Latitude', 'Longitude'], axis=1, errors='ignore', inplace=True)

# Rename columns
df.rename({'City Code': 'City', 'Order Priority Code': 'OrderPriority'}, axis=1, inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Fill missing sales values using median
df.fillna(df.median(numeric_only=True), inplace=True)

# Replace city names
city_map = {1: 'Karachi', 2: 'Lahore', 3: 'Islamabad', 4: 'Faisalabad', 5: 'Quetta', 6: 'Peshawar'}
df['City'] = df['City'].replace(city_map)

df.head()
```

## Top 5 Best Selling Products

```python
top_products = df.groupby('product')['total'].sum().sort_values(ascending=False).head(5)
plt.figure()
sns.barplot(x=top_products.index, y=top_products.values)
plt.title("Top 5 Best Selling Products")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

## Payment Method Distribution

```python
plt.figure()
df['payment_method'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Payment Method Distribution")
plt.show()
```

## Monthwise Sales

```python
plt.figure()
df.groupby('Month')['total'].sum().plot(kind='line')
plt.title("Monthwise Sales")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

## City Wise Sales

```python
plt.figure()
city_sales = df.groupby('City')['total'].sum()
sns.barplot(y=city_sales.index, x=city_sales.values)
plt.title("City Wise Sales")
plt.xlabel("Total Sales")
plt.ylabel("City")
plt.show()
```

## Monthly Sales by Category

```python
plt.figure()
sns.barplot(x='Month', y='total', hue='category', data=df)
plt.title("Monthly Sales by Category")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

## Conclusion

This analysis provides key insights into product performance, sales trends, payment preferences, and geographical performance to improve business decisions.
