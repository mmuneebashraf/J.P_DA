```python
topFiveSales = df[['price', 'quantity', 'Sales']].sort_values(by='Sales', ascending=False).head(5)
print(topFiveSales)
```

```python
topFiveSales = df.sort_values(by='Sales', ascending=False).head(5)
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(data=topFiveSales, x='product', y='Sales', palette='viridis')
plt.title('Top 5 Best Selling Products')
plt.xlabel('Product')
plt.ylabel('Sales')
for p in ax.patches:
    value = p.get_height()                  
    x = p.get_x() + p.get_width() / 2       
    y = value                               

    ax.text(x, y + 5000,                    
            f'{int(value)}',                
            ha='center', fontsize=12, fontweight='bold')
plt.show()
```

## Observations:
The graph shows that “Headphones” and “Bags” are the highest-selling products, with sales around 400,000 each. “Mobiles” and “Shoes” have comparatively lower sales, approximately 320,000 each. Overall, among the top 5 products, “Headphones” hold a slightly leading position, while “Mobiles” and “Shoes” are mid-range performers. The trend also indicates that consumer preference leans more towards electronics and accessories rather than footwear.

```python
payment_counts = df['payment_method'].value_counts()
print(payment_counts)
```

```python
plt.figure(figsize=(8,8))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title("Payment Method Distribution")
plt.show()
```

## Observation:
The pie chart shows that customers use different payment methods with a fairly balanced distribution. JazzCash is the most preferred option, followed closely by Credit Card and Easypaisa. COD also makes up a notable share, showing that many users still prefer paying upon delivery. Debit Card is the least used method, indicating lower preference compared to digital wallets and credit cards. Overall, digital payment methods dominate the distribution.


```python
monthwise_sales = df.groupby('Month')['Sales'].sum()
print(monthwise_sales)
```

```python
plt.figure(figsize=(10,6))
plt.plot(monthwise_sales.index, monthwise_sales.values, marker='o', linestyle='-', color='b')
plt.title('Monthwise Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.xticks(rotation=45) 
plt.show()
```

## Observation:
The line graph shows that monthly sales fluctuate significantly throughout the year. The highest sales are recorded in March, followed by strong performance in April, January, and May. In contrast, August, July, and September show noticeably lower sales levels. Overall, the trend indicates alternating periods of high and low sales rather than a consistent upward or downward pattern. This suggests that sales may be influenced by seasonal or external factors.

```python
df['City'] = df['City'].replace({
    'New york': 'New York',
    'LA': 'Los Angeles',
    'San fran': 'San Francisco'
})
print(df['City'].unique())
```

```python
city_sales = df.groupby('City')['Sales'].sum().sort_values()
city_sales
```

```python
city_names = ["Karachi", "Lahore", "Islamabad", "Faisalabad", "Quetta", "Peshawar"]

plt.figure(figsize=(10,6))
ax = sns.barplot(x=city_sales.index, y=city_sales.values, palette='viridis')
plt.title('City-wise Sales')
plt.xlabel('City')
plt.ylabel('Total Sales')

for i, value in enumerate(city_sales.values):
    ax.text(i,max(city_sales.values) + (max(city_sales.values) * 0.03),  
        city_names[i],
        ha='center',
        va='bottom',
        fontsize=12,
        fontweight='bold',
        rotation=0) 
    
plt.ylim(0, max(city_sales.values) * 1.15)   # space upar add kar diya
plt.show()
```

## Observations:
The bar graph compares total sales across six cities. City 2 shows the highest sales, making it the strongest-performing location. City 5 also performs well but is still lower than City 2. Cities 1, 4, and 6 have the lowest sales, indicating weaker market activity in those areas. City 3 remains in the middle range with moderate sales. Overall, the graph shows that sales are uneven, with some cities performing much better than others.

```python
monthly_category_sales = df.pivot_table(index='Month', columns='category', values='Sales', aggfunc='sum')
month_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
monthly_category_sales = monthly_category_sales.reindex(month_order)
```

```python
months = monthly_category_sales.index
categories = monthly_category_sales.columns
num_months = len(months)
num_categories = len(categories)

bar_width = 0.8 / num_categories  
positions = range(num_months)

plt.figure(figsize=(12,6))

for i, category in enumerate(categories):
    plt.bar([p + i*bar_width for p in positions], monthly_category_sales[category], width=bar_width, label=category)

plt.xticks([p + bar_width*(num_categories-1)/2 for p in positions], months)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales by Category')
plt.legend()
plt.show()
```

## Observations:
The bar chart shows monthly sales for Accessories, Electronics, and Fashion. Accessories have high sales in January and March, but drop mid-year. Electronics sales are relatively stable, with peaks in March, May, and October. Fashion sales fluctuate, performing well in February, May, and December, but dropping in mid-year. Overall, March and May are strong months, while June to September show lower sales across categories.

## Overall Observation:
The analysis indicates that a small number of products contribute most of the sales, with Electronics and Fashion being the top sellers. Credit Card is the most commonly used payment method, whereas Easypaisa and JazzCash are used less frequently. Among the cities, Karachi and Lahore lead in sales. Sales reach their highest in December, showing a seasonal trend. These findings can help in planning inventory, marketing strategies, and payment options to boost revenue and enhance customer satisfaction.

## Conclusion:
The analysis shows that Headphones and Bags are the best-performing products, highlighting strong customer interest in electronics and accessories. Mobiles and Shoes fall in the average range, indicating moderate demand. Payment trends reveal that JazzCash, Credit Card, and Easypaisa are the most preferred methods, showing a clear shift toward digital payments, while COD also maintains a notable share.

City-wise results indicate that City 2 leads in total sales, while Cities 1, 4, and 6 perform poorly.
