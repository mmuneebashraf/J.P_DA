```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("cars.csv")
print("Data loaded successfully!")
print(df.info(), "\n")
```

#### Check Duplicate Records

```python
duplicates = df.duplicated().sum()
print(f"Number of duplicate records: {duplicates}")
```

#### Drop Duplicate Records

```python
df = df.drop_duplicates()
print("Duplicates dropped.\n")
```

#### Check missing or null values

```python
print("Missing values per column:\n", df.isnull().sum(), "\n")
```

#### Drop missing or null values

```python
df_clean = df.dropna()
print("Missing/null values dropped.\n")
```

#### Show top 10 average MSRP (Price) per car brand (rounded figure)?

```python
top10_avg_price = (
    df_clean.groupby("Make")["MSRP"]
    .mean()
    .round()
    .sort_values(ascending=False)
    .head(10)
)
print("Top 10 Average MSRP per Car Brand:\n", top10_avg_price, "\n")
```

#### Show the average MSRP (price) per years (rounded figure)?

```python
avg_price_per_year = df_clean.groupby("Year")["MSRP"].mean().round()
print("Average MSRP per Year:\n", avg_price_per_year, "\n")
```

#### Show average Engine HP (power) rounded figure and MPG (efficiency) rounded figure per year?

Hint: Average of (city mpg + highway MPG)

```python
df_clean["Average_MPG"] = ((df_clean["city mpg"] + df_clean["highway MPG"]) / 2).round()

avg_hp_mpg = (
    df_clean.groupby("Year")[["Engine HP", "Average_MPG"]]
    .mean()
    .round()
)
print("Average Engine HP and MPG per Year:\n", avg_hp_mpg, "\n")
```

#### What percentage of cars are Compact, Midsize, or Large?

```python
style_counts = df_clean["Vehicle Size"].value_counts(normalize=True) * 100
compact_mid_large = style_counts.loc[["Compact", "Midsize", "Large"]]
print("Percentage of Compact, Midsize, or Large cars:\n", compact_mid_large, "\n")

```

#### Which vehicle styles are the most sold (most common) in the dataset?

```python
most_common_style = df_clean["Vehicle Style"].value_counts().head(5)
print("Most common vehicle styles:\n", most_common_style, "\n")
```

#### Which car brands have the most vehicles priced at $40,000 or more and with a popularity score of at least 1000?

```python
high_price_popular = df_clean[
    (df_clean["MSRP"] >= 40000) & (df_clean["Popularity"] >= 1000)
]
brands_high_price = high_price_popular["Make"].value_counts().head(10)
print("Brands with vehicles priced ≥ $40,000 and popularity ≥ 1000:\n", brands_high_price, "\n")
```

#### Check if there are outliers in Engine HP show boxplot

```python
plt.boxplot(df_clean["Engine HP"].dropna())
plt.title("Boxplot of Engine HP")
plt.show()
```

#### Check if there are outliers in Engine Cylinder show boxplot

```python
plt.boxplot(df_clean["Engine Cylinders"].dropna())
plt.title("Boxplot of Engine Cylinders")
plt.show()
```

#### Find upper bound, lower bound, number of outliers in Engine HP

```python
Q1 = df_clean["Engine HP"].quantile(0.25)
Q3 = df_clean["Engine HP"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df_clean[
    (df_clean["Engine HP"] < lower_bound) | (df_clean["Engine HP"] > upper_bound)
]

print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Number of outliers in Engine HP: {len(outliers)}\n")
```

#### Fill missing values in Number of Doors column with median value then check if values are filled

```python
median_doors = df["Number of Doors"].median()
df["Number of Doors"].fillna(median_doors, inplace=True)

print("Missing values in 'Number of Doors' filled with median.")
print(df["Number of Doors"].isnull().sum(), "missing values remain.\n")
```
