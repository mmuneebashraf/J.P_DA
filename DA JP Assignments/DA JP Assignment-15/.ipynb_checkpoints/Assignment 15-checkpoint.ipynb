#### **1. What is the average TotalPayBenefits for each JobTitle? exclude ('Not Provided', 'Not provided')**

```python
import sqlite3
import pandas as pd
```

```python
db_path = "database.sqlite"
conn = sqlite3.connect(db_path)
```

```python
query1 = """
SELECT JobTitle, 
       AVG(TotalPayBenefits) AS AvgTotalPayBenefits
FROM Salaries
WHERE JobTitle NOT IN ('Not Provided', 'Not provided')
GROUP BY JobTitle
ORDER BY AvgTotalPayBenefits DESC;
"""
df1 = pd.read_sql(query1, conn)
print("Average TotalPayBenefits by JobTitle:\n", df1)
```

#### **2. Show The JobTitle and Overtime Pay for All Employees With Overtime Pay greater than 50000**

```python
query2 = """
SELECT EmployeeName, JobTitle, OvertimePay
FROM Salaries
WHERE OvertimePay > 50000;
"""
df2 = pd.read_sql(query2, conn)
print("Employees with OvertimePay > 50000:\n", df2)
```

#### **3. What is the total number of unique employees? exclude ('Not Provided', 'Not provided')**

```python
query3 = """
SELECT COUNT(DISTINCT EmployeeName) AS TotalUniqueEmployees
FROM Salaries
WHERE EmployeeName NOT IN ('Not Provided', 'Not provided');
"""
df3 = pd.read_sql(query3, conn)
print("Total Unique Employees:\n", df3)
```

#### **4.Show The Top 10 highest Paid Employees**

```python
query4 = """
SELECT EmployeeName, JobTitle, TotalPayBenefits
FROM Salaries
ORDER BY TotalPayBenefits DESC
LIMIT 10;
"""
df4 = pd.read_sql(query4, conn)
print("Top 10 Highest Paid Employees:\n", df4)
```

#### **5.Show All Employees Ordered By Their TotalPay Benefits In Descending Order**

```python
query5 = """
SELECT EmployeeName, JobTitle, TotalPayBenefits
FROM Salaries
ORDER BY TotalPayBenefits DESC;
"""
df5 = pd.read_sql(query5, conn)
print("All Employees Ordered By Salary:\n", df5)
```
