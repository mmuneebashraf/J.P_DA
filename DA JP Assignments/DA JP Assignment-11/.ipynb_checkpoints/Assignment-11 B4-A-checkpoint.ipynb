```python
import csv
```

```python
open_file = open('student data.csv', mode='r')
open_file
```

```python
reader = csv.reader(open_file)
reader
```

```python
data = list(reader)
data
```

```python
for per in data[1:]:  
    percentage = float(per[5])

    if percentage>100:
        per.append("More Than 100 % Value Error")
    elif percentage >= 80 and percentage<= 100:
        per.append('A')
    elif percentage >= 70 and percentage<80:
        per.append('B')
    elif percentage >= 60 and percentage<=70:
        per.append('C')
    elif percentage >= 50 and percentage<=60:
        per.append('D')
    elif percentage >= 0 and percentage<=50:
        per.append('F')
    elif percentage <0:
        per.append('-ve Value Error')
```
